Start by investigating the database stored in lego.db and joining the tables into a unified dataset!

> **Hint:** use this SQL query to preview the tables in an unknown database:
```sql
SELECT name FROM sqlite_master
            WHERE type='table'
            ORDER BY name;
```

````python
import sqlite3
import pandas as pd
````

````python
con = sqlite3.connect('lego.db')
cur = con.cursor()
tables = cur.execute("""SELECT name FROM sqlite_master
            WHERE type='table'
            ORDER BY name;""").fetchall()
for table in tables:
    column = table[0]
    data = cur.execute("""SELECT * FROM {};""".format(column)).fetchall()
    df = pd.DataFrame(data)
    df.columns = [i[0] for i in cur.description]
df
````

````python
tables
````

````python
new_result = cur.execute("""SELECT * FROM product_info
                  JOIN product_details
                  USING(prod_id)
                  JOIN product_pricing
                  USING(prod_id)
                  JOIN product_reviews
                  USING(prod_id);""").fetchall()
df_joined = pd.DataFrame(new_result)
df_joined.columns = [j[0] for j in cur.description]
df_joined
````

````python
df_joined.to_csv('lego_joined.csv', index=False)
````
